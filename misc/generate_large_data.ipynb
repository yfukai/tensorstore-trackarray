```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import disk

# Initialize a blank 20000 x 20000 image
image_size = 20000
large_image = np.zeros((image_size, image_size), dtype=np.uint8)

# Function to draw circular blobs
def draw_circular_blobs(image, num_blobs, blob_diameter):
    """Draw circular blobs on the image."""
    radius = blob_diameter // 2
    for _ in range(num_blobs):
        # Randomly select the center of the circle
        center_x = np.random.randint(radius, image.shape[0] - radius)
        center_y = np.random.randint(radius, image.shape[1] - radius)

        # Get the coordinates of the circle
        rr, cc = disk((center_x, center_y), radius, shape=image.shape)

        # Draw the circle
        image[rr, cc] = 255

# Parameters for the blobs
num_blobs = 1000
blob_diameter = 30

# Draw the blobs
draw_circular_blobs(large_image, num_blobs, blob_diameter)

# Display a small portion of the image for verification
plt.imshow(large_image)  # Display top-left corner
plt.title("Top-left corner of the large image with blobs")
plt.colorbar()
plt.show()
```
